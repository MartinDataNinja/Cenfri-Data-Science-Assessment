```python
import pandas as pd
traffic_data = pd.read_excel(r"C:\Users\GK TECH\Downloads\Traffic_Incident_Reports_Clean.xlsx")
print(traffic_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data Validation

traffic_data['Published_Date'] = pd.to_datetime(traffic_data['Published_Date'])
traffic_data['Status_Date'] = pd.to_datetime(traffic_data['Status_Date'])
traffic_data['Issue_Reported'] = traffic_data['Issue_Reported'].astype('category')
traffic_data['Status'] = traffic_data['Status'].astype('category')

traffic_data['Published_Year'] = traffic_data['Published_Date'].dt.year
traffic_data['Published_Month'] = traffic_data['Published_Date'].dt.month
traffic_data['Status_Year'] = traffic_data['Status_Date'].dt.year
traffic_data['Status_Month'] = traffic_data['Status_Date'].dt.month

traffic_data.head()
```

```python
#Heat Map for Traffic Incidences montly

heatmap_data = traffic_data.pivot_table(index='Published_Year', columns='Published_Month', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu', cbar_kws={'label': 'Number of Incidents'})
plt.xlabel('Month')
plt.ylabel('Year')
plt.tight_layout()
plt.savefig(r"C:\Users\GK TECH\Downloads\incidents_by_year_month_heatmap.pdf", format='pdf')
plt.show()
```
